```python
# Group the data by week and meal
df_grouped = df_train.groupby(['week', 'meal_id'])

# Calculate the number of orders for each week and meal
df_counts = df_grouped['num_orders'].sum()

# Unstack the data frame to create a pivot table
df_pivot_table = df_counts.unstack(fill_value=0)

# Create a line plot of the number of orders for each week and meal
plt.figure()
for meal in df_pivot_table.columns:
    plt.plot(df_pivot_table.index, df_pivot_table[meal], label=meal)

plt.xlabel('Week')
plt.ylabel('Number of Orders')
plt.title('Number of Orders for Different Meals by Week')
plt.legend()
plt.show()
```
